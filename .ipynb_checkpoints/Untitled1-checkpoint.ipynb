```python
# for i in range(1,16):
#     data.at[i*30, 'radius_mean']  = 'NULL'
# data.radius_mean.head(40)
for i in range (0,16):
    i = i*30
#     data = data.radius_mean.replace([data.radius_mean[i*30]], 'Null')
#     data['radius_mean'].mask(data['radius_mean'] == data.radius_mean[i*30],'Null', inplace=True)
    data.radius_mean.replace(data.radius_mean[(i)], np.nan, inplace = True)
```

```python
for i in range (0,16):
#     data = data.radius_mean.replace([data.radius_mean[i*30]], 'Null')
#     data['radius_mean'].mask(data['radius_mean'] == data.radius_mean[i*30],'Null', inplace=True)
    i = i*30
    data.area_mean.replace(data.area_mean.at[i], np.nan, inplace = True)
```

```python
for i in range (0,len(data.perimeter_mean)):
    try:
        if data.perimeter_mean.at[i] > 90 and data.perimeter_mean.at[i] <120:
            data.perimeter_mean.replace(data.perimeter_mean.at[i], 'Between 90 and 120', inplace = True)
            
        elif data.perimeter_mean.at[i] > 120:
            data.perimeter_mean.replace(data.perimeter_mean.at[i], 'Greater than 120', inplace = True)
            
        else :
            data.perimeter_mean.replace(data.perimeter_mean.at[i], 'Less than 90', inplace = True)
            
    except : pass
    
```
