# Training the Model 

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.feature_selection import SelectKBest,chi2
from sklearn.tree import DecisionTreeClassifier
```

```python
# Step 1 -> train/test/split
X_train,X_test,y_train,y_test = train_test_split(X[[ 'radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean']],
                                                 X['diagnosis'],
                                                 test_size=0.2,
                                                random_state=2)
```

```python
X_train.shape
```

```python
# imputation transformer
trf1 = ColumnTransformer([
    ('impute_rad_ar',SimpleImputer(),[0,3])
                         ],
    remainder='passthrough')
```

```python
# one hot encoding
trf2 = ColumnTransformer([
    ('ohe_sex_embarked',OneHotEncoder(sparse=False,handle_unknown='ignore'),[3])#after imputation the imputed column comes first
],remainder='passthrough')
```

```python
# Scaling
trf3 = ColumnTransformer([
    ('scale',MinMaxScaler(),slice(0,32))  # no. of column increased due to ohe
])
```

```python

# Feature selection
trf4 = SelectKBest(score_func=chi2,k=7)
```

```python
# train the model
trf5 = DecisionTreeClassifier()
```

```python
pipe = Pipeline([
    ('trf1',trf1),
    ('trf2',trf2),
    ('trf3',trf3),
    ('trf4',trf4),
    ('trf5',trf5)
])
```

```python
# train
pipe.fit(X_train,y_train)
```

```python
# Predict
y_pred = pipe.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
y_test.value_counts()
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

### model validation

```python
from sklearn.model_selection import KFold, cross_val_score
```

```python
k_folds = KFold(n_splits = 5)

scores = cross_val_score(pipe,X_train, y_train, cv = k_folds)
```

```python
print("Cross Validation Scores: ", scores)
print("Average CV Score: ", scores.mean())
print("Number of CV Scores used in Average: ", len(scores))
```
