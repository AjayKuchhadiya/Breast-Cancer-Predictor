# Training the Model 

```python

from sklearn.model_selection import train_test_split

# Obtain target and predictors
y = X.diagnosis         # target data

features = ['radius_mean', 'texture_mean',
        'area_mean', 'smoothness_mean', 'compactness_mean','perimeter_mean', 'concavity_mean',
        'concave points_mean', 'symmetry_mean']

X = X[features].copy()  # feature data

# Breaking off validation set from training data
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2,
                                                      random_state=0)
```

```python
X_train.head()
```

### working with categorical values

```python
# Replacing catigorical values from the dataset 

low_cardinality_cols = [cname for cname in X_train.columns if X_train[cname].dtype == 'object'
                        ]
```

```python
from sklearn.preprocessing import OrdinalEncoder

ordinal_encoder = OrdinalEncoder()
X_train[low_cardinality_cols] = ordinal_encoder.fit_transform(X_train[low_cardinality_cols])
X_valid[low_cardinality_cols] = ordinal_encoder.transform(X_valid[low_cardinality_cols])
```

```python
X_train.head()
```

### removing null values

```python
from sklearn.impute import SimpleImputer

# imputation
my_impute = SimpleImputer(strategy = 'mean') # Imputing mean inplace of null values
imputed_X_train = pd.DataFrame(my_impute.fit_transform(X_train))
imputed_X_valid =pd.DataFrame(my_impute.transform(X_valid))

# imputation removed column names; put them back
imputed_X_train.columns = X_train.columns
imputed_X_valid.columns = X_valid.columns
```

```python
imputed_X_train.isnull().any()  # Null values removed
```

```python
imputed_X_train.info()
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
final_model = RandomForestRegressor()
final_model.fit(imputed_X_train, y_train)
```

```python
predictions = final_model.predict(imputed_X_valid)
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(predictions, y_valid)
print('MAE value is :',mae)
```

```python
# 0.0882 with fractal
# 0.0853 without fractal
```

### model validation

```python
from sklearn.model_selection import KFold, cross_val_score
```

```python
k_folds = KFold(n_splits = 5)

scores = cross_val_score(final_model, imputed_X_train, y_train, cv = k_folds)
```

```python
print("Cross Validation Scores: ", scores)
print("Average CV Score: ", scores.mean())
print("Number of CV Scores used in Average: ", len(scores))
```
